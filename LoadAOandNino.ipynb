```python
import pandas as pd
import numpy as np
import xarray as xr
```

### Load ao.csv

```python
path='/home/ubuntu/clim-data/'
file='ao.data'
index_name='ao'

# Read the dataframe
df=pd.read_table(path+file,skiprows=1,
                 header=None,delim_whitespace=True,
                 index_col=0,parse_dates=True,
                 na_values=['-99.9','-99.90'])

df.drop(df.tail(3).index,inplace=True)

# Convert to xarray
start_date=str(df.index[0])+'-'+str(df.columns[0])+'-01'
end_date=str(df.index[-1])+'-'+str(df.columns[-1])+'-01'

dates=pd.date_range(start=start_date,end=end_date,freq='MS')
clim_index=df.T.unstack().astype('float').values

clim_index

ds_index=xr.DataArray(clim_index,
                      coords={'time':dates},
                              dims=['time'])        

ds_index
ds_index=ds_index.to_dataset(name=index_name)

ds_index=ds_index.sel(time=slice('2000-01-01','2016-12-01'))

ds_index.to_netcdf('/home/ubuntu/indexes/ao.nc')
ds_index
```
