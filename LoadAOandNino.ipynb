### Load Nino34

```python
import xarray as xr
import cftime
import pandas as pd


file='/home/ubuntu/iersst_nino3.4a.nc'
ds = xr.open_dataset(file,decode_times=False)#.rename({'WEDCEN2':'time'})


units, reference_date = ds.time.attrs['units'].split('since')
ds['time'] = pd.date_range(start=reference_date, periods=ds.sizes['time'], freq='MS')

ds=ds.sel(time=slice('2000-01-01','2016-12-01'))
ds=ds.rename({'Nino3.4': 'sst'})
ds.to_netcdf('/home/ubuntu/indexes/nino34.nc')

```
